```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "x":[0, 1, 2, 3, 3, 5, 5, 6, 7, 7, 10, 5],
    "y":[96, 85, 82, 74, 95, 68, 76, 58, 65, 75, 50, 84]
}
```

```python
# We will calculate slope y = mx + c
```

```python
# no of observation
n = len(data["x"])
n
```

```python
# m = [n(∑xy) - (∑x)(∑y)] / [n(∑x²) - (∑x)²]
# Intercept (c) = (∑y - m(∑x)) / n
```

```python
xy = [data["x"][i] * data["y"][i] for i in range(n)]
sum_xy = sum(xy)
```

```python
sum_x = sum(data["x"])
sum_y = sum(data["y"])
squared_x = [x**2 for x in data["x"]]
sum_squared_x = sum(squared_x)
```

```python
m = (n*sum_xy - sum_x * sum_y)/ (n*sum_squared_x - sum(data["x"])**2)
m
```

```python
c = (sum_y - m * sum_x)/n
c
```

```python
user_hrs = int(input("Enter No. of hrs TV watched : "))
```

```python
prediction = m * user_hrs + c
prediction
```

```python
# plot slope
x = np.linspace(0, 12, 100)
y = m * x + c

plt.plot(x, y, label=f'y = {m}x + {c}', color='red')
plt.scatter(data["x"], data["y"])
```
